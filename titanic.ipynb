```python
import numpy as np
import sklearn
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_train.head()
```

```python
y_train = df_train['Survived']
x_train = df_train.drop('Survived',axis = 1)
x_test = df_test
x_test.head()
```

```python
#data cleaning
x_train = x_train.drop('PassengerId',axis=1)
x_test = x_test.drop('PassengerId',axis=1)
```

```python
x_train = x_train.drop('Name',axis=1)
x_test = x_test.drop('Name',axis=1)
```

```python
x_train = x_train.drop('Cabin',axis=1)
x_test = x_test.drop('Cabin',axis=1)
```

```python
x_train = x_train.drop('Ticket',axis=1)
x_test = x_test.drop('Ticket',axis=1)
```

```python
x_train.isnull().sum()
```

```python
x_train['Sex']=pd.get_dummies(x_train['Sex'],drop_first=True)
x_train.head()
```

```python
x_test['Sex']=pd.get_dummies(x_test['Sex'],drop_first=True)
x_test.head()
```

```python
x_train['Embarked'] = x_train['Embarked'].fillna('S')
x_train['Embarked'].unique()
```

```python
x_train['Embarked']=pd.get_dummies(x_train['Embarked'],drop_first=True)
x_train.head()
```

```python
x_train['Pclass']=pd.get_dummies(x_train['Pclass'],drop_first=True)
x_train.head()
```

```python
x_test['Pclass']=pd.get_dummies(x_test['Pclass'],drop_first=True)
x_test.head()
```

```python
x_test['Embarked']=pd.get_dummies(x_test['Embarked'],drop_first=True)
x_test.head()
```

```python
x_train['Age']
```

```python
avg_age = x_train['Age'].mean()
x_train['Age']=x_train['Age'].fillna(avg_age)
x_train['Age'][888]
```

```python
avg_age_test = x_test['Age'].mean()
x_test['Age']=x_test['Age'].fillna(avg_age_test)
```

```python
x_test['Fare'] = x_test['Fare'].fillna(x_test['Fare'].mean())
x_test.head()
```

```python
#normalization:
```

```python
x_test['Age'] = x_test['Age']/x_train['Age'].max()
x_train['Age'] = x_train['Age']/x_train['Age'].max()
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_test['Fare'] = x_test['Fare']/x_train['Fare'].max()
x_train['Fare'] = x_train['Fare']/x_train['Fare'].max()
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(max_iter=1000)
```

```python
clf.fit(x_train,y_train)
```

```python
result = clf.predict(x_test)
result
```
